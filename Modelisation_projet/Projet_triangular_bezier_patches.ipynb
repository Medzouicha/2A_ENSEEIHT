## DeCasteljau Evaluation

The surface is desribed by : 
$$
S(u, v, w)=\sum_{\substack{i+j+k=n \\ i, k=0}} B_{i, j, k}^n(u, v, w) p_{i, j, k,}
$$
and $B_{i, k, k}^n$ are the bivariate Bernstein polynomials, which are given by
$$
B_{i, j, k}^n(u, v, w)=\frac{n !}{i ! j ! k !} u^i v^j w^k
$$
